```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sale = {'Audi':124, 'BMW':98, 'Mercedes':113}
```

```python
Frequency = list(sale.keys())
Car = list(sale.values())
```

```python
fig = plt.figure(figsize = (10,5))
```

```python
c = ['blue', 'green', 'red']
```

```python
plt.bar(Frequency, Car, color = c)
plt.title("Sales")
plt.xlabel("Cars")
plt.ylabel("Frequency")
plt.show()
```

```python
Cars = ['Audi', 'BMW', 'Mercedes']
data = [124, 98, 113]
```

```python
fig = plt.figure(figsize= (10 ,15))
```

```python
plt.pie(data, labels = Cars, startangle=0, 
         shadow=True,colors = ["blue", "green", "red"], autopct="%2.1f%%")
# explode = [0.2,0.1,0.12]
plt.legend(title = 'Car', loc= 4)
plt.show()
```
